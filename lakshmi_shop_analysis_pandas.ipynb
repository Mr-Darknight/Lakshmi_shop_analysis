```python
import pandas as pd
df = pd.read_excel('C:\\Users\\laksh\\OneDrive\\Desktop\\Lakshmi Traders\\realfile_prchase_data.xlsx')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst=True,format='mixed',errors='coerce')
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('-','_')
df.head()
```

```python
df.groupby('category')['total_cost'].sum() # total purchase category wise.
```

```python
df.groupby('supplier')['total_cost'].sum() # total purchase supplier wise
```

```python
df.insert(7,'total_quantity', df['unit_per_pack']*df['pack_count']*df['unit_size']) # Add a column for total quantity of a item 
df.head()
```

```python
df.insert(9,'cost_per_unit',round(df['total_cost']/df['total_quantity'],2)) # Add a column to find per unit price
df
```

```python
df.insert(11,'month',df['date'].dt.strftime('%b'))
df.head(10)
```

```python
Q1 = df['cost_per_unit'].quantile(0.25)
Q3 = df['cost_per_unit'].quantile(0.75)
IQR = Q3-Q1
```

```python
df_clean = df[
    (df['cost_per_unit']>= Q1 - 1.5*IQR)&
    (df['cost_per_unit']<= Q3 + 1.5*IQR)
]
```

```python
median_rate = df.groupby(['item_name','unit_size','pack_type'])['cost_per_unit'].median()
df["median_rate"] = df.set_index(
    ["item_name","unit_size","pack_type"]
).index.map(median_rate)
df[['item_name','cost_per_unit','median_rate']]
```

```python
df['rate_deviation'] = round(df['cost_per_unit']/df['median_rate'],2)
df
```

```python
def label(x):
    if x > 1.1:
        return "Overvalued"
    elif x < 0.9:
        return "Undervalued"
    else:
        return "Normal"

df["rate_status"] = df["rate_deviation"].apply(label)
```
